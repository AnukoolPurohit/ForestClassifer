```python
class PlanetClassifer(nn.Module):
    def __init__(self, arch, output=2):
        super().__init__()
        model = arch(pretrained=True)
        in_features = model.fc.in_features
        linear = nn.Sequential(nn.Linear(in_features, 256),
                                 nn.ReLU(),
                                 nn.Dropout(0.2),
                                 nn.Linear(256, output))
        model.fc = linear
        self.model = model.cuda()
        self.freeze()
        return
    
    def forward(self, x):
        return self.model(x)
    
    def freeze(self):
        children = [chlid for chlid in self.model.children()]
        for child in children[0:8]:
            for params in child.parameters():
                params.requires_grad = False
        return
    
    def unfreeze(self):
        for params in self.model.parameters():
            params.requires_grad = True
        return
```

```python
class Logger():
    def __init__(self):
        self.data = {
            'training-loss': [],
            'validation-loss': [],
            'learning-rate': []
        }
        return
    
    def log_data(self, data):
        for key in self.data.keys():
            self.data[key].extend(data[key])
        return
    
    def los_lists(self, listdata):
        data = {
            'training-loss': listdata[0],
            'validation-loss': listdata[1],
            'learning-rate': listdata[2]
        }
        self.log_data(data)
        return
    
    def plot_loss(self):
        training_loss = self.data['training-loss']
        validation_loss = self.data['validation-loss']
        
        ax1 = plt.subplot(2,1,1)
        ax1.plot(training_loss, 'b', label='Training Loss')
        ax1.legend(loc='best')
        ax2 = plt.subplot(2,1,2)
        ax2.plot(validation_loss, 'y', label='Validation Loss')
        plt.legend(loc='best')
        plt.show()
        
```

```python
class NNTrainer():
    def __init__(self, data, arch):
        self.data = data
        self.model = PlanetClassifer(arch, output=data.c)
        self.loss_func = nn.BCEWithLogitsLoss()
        self.log = Logger()
        return
    
    @staticmethod
    def update(x,y,model,loss_func,lr):
        model.train()
        opt = optim.Adam(model.parameters(), lr)
        y_hat = model(x)
        loss = loss_func(y_hat, y)
        loss.backward()
        opt.step()
        opt.zero_grad()
        return loss.item()
    
    @staticmethod
    def evaluate(x,y,model,loss_func):
        model.eval()
        y_hat = model(x)
        loss = loss_func(y_hat, y)
        return loss.item()
    
    def freeze(self):
        self.model.freeze()
        return
    
    def unfreeze(self):
        self.model.unfreeze()
        return
    
    def train(self,lr,epochs=3):
        training_errors=[]
        validation_errors = []
        print_cols = False
        bar = tqdm_notebook(range(epochs), leave=False)
        for i in bar:
            res = self.train_epoch(lr)
            self.log.los_lists(res)
            train_error = (sum(res[0])/len(res[0]))
            valid_error = (sum(res[1])/len(res[1]))
            
            training_errors.append(train_error)
            validation_errors.append(valid_error)
            
            metalist = ["Training Loss", "Validation Loss"]
            row_format ="{:>25}" * (len(metalist) + 1)
            if print_cols == False:
                print(row_format.format("Epoch", *metalist))
                print_cols = True
            print (row_format.format(i+1, "%.6f"%train_error, "%.6f"%valid_error))
            
        return
            
    def train_epoch(self,lr):
        train_losses = []
        valid_losses = []
        lr_log = []
        train_iter = tqdm_notebook(self.data.train_dl, leave=False)

        self.model.train()
        for x,y in train_iter:
            loss = self.update(x,y,self.model,self.loss_func,lr)
            train_iter.set_postfix({'Loss'  : loss})
            train_losses.append(loss)
            lr_log.append(lr)
        valid_iter = tqdm_notebook(self.data.valid_dl, leave=False)
        self.model.eval()
        for x,y in valid_iter:
            loss = self.evaluate(x,y,self.model,self.loss_func)
            valid_iter.set_postfix({'Loss':loss})
            valid_losses.append(loss)
        return [train_losses, valid_losses, lr_log]
    
    def save(self, filename):
        picklefile = open(filename, 'wb')
        pickle.dump(self.model, picklefile)
        picklefile.close()
        return
    
    def load(self, filename):
        picklefile = open(filename, 'rb')
        self.model = pickle.load(picklefile)
        picklefile.close()
        return
    
    def predict_batch(self, batch):
        self.model.eval()
        res = self.model(batch)
        return res
    
    def predict_test(self):
        res = []
        for x,_ in tqdm(self.data.test_dl):
            res.append(self.predict_batch(x))
        res = torch.cat(res)
        return nn.Sigmoid()(res)
```

```python
def get_df(preds, data, thresh=0.2):
    labeled_preds = [' '.join([data.train_ds.classes[i] for i,p in enumerate(pred) if p > thresh]) for pred in preds]
    fnames = data.test_ds.x.items
    df = pd.DataFrame({'image_name': fnames, 'tags': labeled_preds}, columns=['image_name', 'tags'])
    df['extra'] = df['image_name'].str.split('_', n=1, expand=True)[1]
    df['extra'] = df['extra'].astype(float)
    df.sort_values(by='extra', inplace=True)
    df.drop(columns='extra', inplace=True)
    return df
```
